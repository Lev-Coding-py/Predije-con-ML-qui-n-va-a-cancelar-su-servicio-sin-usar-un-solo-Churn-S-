```python
cols1=['SeniorCitizen','tenure','TotalCharges','Contract_Month-to-month','Contract_One year','Contract_Two year','PaymentMethod_Bank transfer (automatic)','PaymentMethod_Credit card (automatic)','PaymentMethod_Electronic check','PaymentMethod_Mailed check']
```

```python
cols2=['SeniorCitizen','tenure','TotalCharges','Contract_Month-to-month','Contract_One year','Contract_Two year','PaymentMethod_Bank transfer (automatic)','PaymentMethod_Credit card (automatic)','PaymentMethod_Electronic check','PaymentMethod_Mailed check','Churn']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
dx_scaled = scaler.fit_transform(dx)                                   
dx = pd.DataFrame(dx_scaled, columns=dx.columns, index=dx.index)        
```

```python
from sklearn.cluster import KMeans

modelo = KMeans(n_clusters=2, random_state=42)
modelo.fit(dx)
```

```python
dl['Cluster']=modelo.fit_predict(dx)
```

```python
dl['Cluster']=modelo.labels_
```

## RECOMENDACIONES

### 1. Estrategia de fidelización
Los clientes con mayor antigüedad y contratos de uno o dos años presentan una mayor estabilidad y representan el segmento de mayor valor para la empresa. En contraste, los clientes con contratos mensuales y menor tiempo de permanencia muestran un comportamiento más inestable y requieren estrategias de retención.

### 2. Automatización de campañas
Implementar campañas automatizadas mediante herramientas como Make o n8n para enviar promociones, descuentos y recordatorios personalizados por correo electrónico o WhatsApp a los clientes con mayor riesgo de abandono. En casos específicos, el sistema puede derivar al cliente a un asesor humano para brindar atención personalizada.

### 3. Asistente basado en IA generativa
Integrar un asistente utilizando modelos de lenguaje como OpenAI, Claude o modelos open source que analice la información del cliente y genere explicaciones en lenguaje natural sobre su nivel de estabilidad, los factores que influyen en su comportamiento y recomendaciones para mejorar su retención.

### 4. Desarrollo de una aplicación
Desarrollar una aplicación web donde usuarios no técnicos puedan ingresar la información de un cliente y obtener automáticamente:
- El segmento al que pertenece.
- Una estimación de su nivel de estabilidad.
- Recomendaciones de retención.
- Explicaciones generadas por IA para facilitar la toma de decisiones.

### 5. Priorización de clientes

No todos los clientes requieren las mismas acciones de retención. Se recomienda priorizar las campañas sobre aquellos segmentos con mayor riesgo de abandono y mayor valor económico, optimizando así los recursos destinados a marketing y atención al cliente.
